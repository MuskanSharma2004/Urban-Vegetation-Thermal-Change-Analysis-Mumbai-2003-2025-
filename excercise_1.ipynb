```python
import geemap
```

```python
ee.Authenticate()
ee.Initialize(project="sona-ee")
```

```python
map = geemap.Map(ee_initialize=False)
map
```

```python
State = ee.FeatureCollection('projects/sona-ee/assets/STATE_BOUNDARY')
map.addLayer(State,{},'India_states')
map
```

```python
ASSAM = State.filter(ee.Filter.eq('STATE','ASSAM'))
map.addLayer(ASSAM,{},'ASSAM')
map
```

```python
DEM = ee.Image("CGIAR/SRTM90_V4")
ASSAM_DEM = DEM.clip(ASSAM)
map.addLayer(ASSAM_DEM,{},'DEM')
map
```
